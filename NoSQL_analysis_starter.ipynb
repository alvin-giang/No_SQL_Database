## Notebook Set Up

```python
# Import dependencies
from pymongo import MongoClient
import pandas as pd
from pprint import pprint
```

```python
# Create an instance of MongoClient
mongo = MongoClient(port=27017)
```

```python
# assign the uk_food database to a variable name
db = mongo['uk_food']
```

```python
# review the collections in our database
print(db.list_collection_names())
```

```python
# assign the collection to a variable
establishments = db['establishments']
```

## Part 3: Exploratory Analysis
Unless otherwise stated, for each question: 
* Use `count_documents` to display the number of documents contained in the result.
* Display the first document in the results using `pprint`.
* Convert the result to a Pandas DataFrame, print the number of rows in the DataFrame, and display the first 10 rows.

### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {'scores.Hygiene': {'$eq':20}}

# Use count_documents to display the number of documents in the result
hygiene_20_documents_count = establishments.count_documents(query)
print(f'Number of documents which has a hygiene score equal to 20 is {hygiene_20_documents_count}')
# Display the first document in the results using pprint
hygiene_results = list(establishments.find(query))
pprint(hygiene_results[0])
```

```python
# Convert the result to a Pandas DataFrame
hygiene_20_df = pd.json_normalize(hygiene_results)
# Display the number of rows in the DataFrame
row_count = hygiene_20_df.shape[0]
print(f'Number of rows is {row_count}')
# Display the first 10 rows of the DataFrame
hygiene_20_df.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query = {'LocalAuthorityName': {"$regex": "London"},
         'RatingValue': {'$gte': 4}}

# Use count_documents to display the number of documents in the result
London_rate_4_count = establishments.count_documents(query)
print(f'Number of documents which is in London and has a Rating value greater than or equal to 4 is {London_rate_4_count}')
# Display the first document in the results using pprint
London_4_results = list(establishments.find(query))
pprint(London_4_results[0])
```

```python
# Convert the result to a Pandas DataFrame
London_4_df = pd.json_normalize(London_4_results)
# Display the number of rows in the DataFrame
row_count = London_4_df.shape[0]
print(f'Number of rows is {row_count}')
# Display the first 10 rows of the DataFrame
London_4_df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

degree_search = 0.01
latitude = establishments.find_one({"BusinessName": "Penang Flavours"})['geocode']['latitude'] 
longitude = establishments.find_one({"BusinessName": "Penang Flavours"})['geocode']['longitude'] 
lat_min = latitude - degree_search
lat_max = latitude + degree_search
long_min = longitude - degree_search
long_max = longitude + degree_search

# Define the query
query = {"RatingValue": 5,
         "geocode.latitude": {"$gte": lat_min, "$lte": lat_max},
         "geocode.longitude": {"$gte": long_min, "$lte": long_max}}

# Sort by hygiene score
sort = [("scores.Hygiene", 1)]

# Limit to top 5 establishments
limit = 5 

# Print the results
top_5_results = list(establishments.find(query).sort(sort).limit(limit))
pprint(top_5_results)
```

```python
# Convert result to Pandas DataFrame
top_5_df = pd.json_normalize(top_5_results)
top_5_df
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that:

# 1. Matches establishments with a hygiene score of 0
match_query = {'$match': {'scores.Hygiene': 0}}

# 2. Groups the matches by Local Authority
group_query = {'$group': {'_id': "$LocalAuthorityName", 'count': { '$sum': 1 }}}

# 3. Sorts the matches from highest to lowest
sort_values = {"$sort": {"count": -1}}

# Put the pipeline together
pipeline = [match_query, group_query, sort_values]

# Print the number of documents in the result
hygiene_0_results = list(establishments.aggregate(pipeline))
print(f'The number of documents in the results is {len(hygiene_0_results)}')
# Print the first 10 results
pprint(hygiene_0_results[0:10])
```

```python
# Convert the result to a Pandas DataFrame
hygiene_0_df = pd.json_normalize(hygiene_0_results)
# Display the number of rows in the DataFrame
row_count = hygiene_0_df.shape[0]
print(f'Number of rows is {row_count}')
# Display the first 10 rows of the DataFrame
hygiene_0_df.head(10)
```
